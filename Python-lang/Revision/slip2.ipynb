Q1.1)Generate and draw a graph G1 with vertex set {‘a’,’b’,’c’,’d’} and edge set {('a', 'd'), ('b','c'),
('b','d'), ('a','c')}, showing labeled vertices and edges

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1 = nx.Graph()
```

```python
vertices={'a','b','c','d'}
edges = {('a', 'd'), ('b','c'),('b','d'), ('a','c')}
```

```python
G1.add_nodes_from(vertices)
G1.add_edges_from(edges)
```

```python
nx.draw(G1, with_labels=True,edge_color='cyan',node_color='')
plt.show()
```
