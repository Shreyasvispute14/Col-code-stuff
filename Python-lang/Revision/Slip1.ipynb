Q1)Generate graph G with vertex(node) set {1,2,3,4,5} and the edge set {(1,5),
(1,3),(1,2),(2,3),(2,4), (3,4), (4,5)} .

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
vertices = {1,2,3,4,5}
G.add_nodes_from(vertices)
```

```python
edges = {(1,5),(1,3),(1,2),(2,3),(2,4), (3,4), (4,5)}
G.add_edges_from(edges)
```

```python
plt.figure(figsize=(8,6))
nx.draw(G,with_labels=True,node_color='lightblue',node_size=700,font_color='black',font_weight='bold',edge_color='gray')
plt.show()
```

Q2)Generate a star graph with 7 vertices and visualize it using NetworkX


```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
n=7
star_graph = nx.star_graph(n-1)
```

```python
plt.figure(figsize=(8,6))
nx.draw(star_graph, with_labels=True, node_color='lightblue', node_size=5000, edge_color='gray')
plt.show()
```

Q3)Draw a null graph with N₁₂ vertices and analyse its properties using python



```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
n=12
null_graph=nx.Graph()
```

```python
null_graph.add_nodes_from(range(1,n+1))
```

```python
plt.figure(figsize=(8,6))
pos = nx.spring_layout(null_graph)
nx.draw(null_graph, with_labels=True, node_color='lightblue', node_size=5000, edge_color='gray')
plt.show()

print("Properties of null graph N12:")
print(f"Number of veritces: {null_graph.number_of_nodes()}")
print(f"Number of edges: {null_graph.number_of_edges()}")
```
