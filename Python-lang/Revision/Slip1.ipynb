Q1.1)Generate graph G with vertex(node) set {1,2,3,4,5} and the edge set {(1,5),
(1,3),(1,2),(2,3),(2,4), (3,4), (4,5)} .

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
vertices = {1,2,3,4,5}
G.add_nodes_from(vertices)
```

```python
edges = {(1,5),(1,3),(1,2),(2,3),(2,4), (3,4), (4,5)}
G.add_edges_from(edges)
```

```python
plt.figure(figsize=(8,6))
nx.draw(G,with_labels=True,node_color='lightblue',node_size=700,font_color='black',font_weight='bold',edge_color='gray')
plt.show()
```

Q1.3)Draw a null graph with N vertices and analyse its properties using python



```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
n=12
null_graph=nx.Graph()
```

```python
null_graph.add_nodes_from(range(1,n+1))
```

```python
plt.figure(figsize=(8,6))
pos = nx.spring_layout(null_graph)
nx.draw(null_graph, with_labels=True, node_color='lightblue', node_size=5000, edge_color='gray')
plt.show()

print("Properties of null graph N12:")
print(f"Number of veritces: {null_graph.number_of_nodes()}")
print(f"Number of edges: {null_graph.number_of_edges()}")
```

Q2.3)Generate a cycle graph C using NetworkX. Find the number of vertices, edges, and degrees
of each vertex.



```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
C8 = nx.cycle_graph(8)
```

```python
num_verticles = C8.number_of_nodes()
num_edges = C8.number_of_edges()
```

```python
degrees = dict(C8.degree())
```

```python
print("Number of verticles (C8):",num_verticles)
print("Number of edges (C8):",num_edges)
print("Degrees of each vertex (C8):",degrees)
```

```python
nx.draw(C8,with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_color='black', font_weight='bold', edge_color='gray')
plt.show()
```

Q3.3)Draw the complete graphs, for example, K5, K30, K45.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_complete_graph(n):
    K_n=nx.complete_graph(n)

    plt.figure(figsize=(8,6))
    nx.draw(K_n,with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_color='black', font_weight='bold', edge_color='gray')
    plt.show()

draw_complete_graph(30)
```
